## 3. Clustering Neighbourhoods in Toronto

```python
# Filtering Toronto Borough.

df_Toronto = df_mer[df_mer['Borough'].str.contains('Toronto',regex=False)]
df_Toronto
```

```python
#Showing all the Neighbourhoods of the above data frame using Folium on Map

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df_Toronto['Latitude'],df_Toronto['Longitude'],df_Toronto['Borough'],df_Toronto['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='gray',
    fill=True,
    fill_color='blue',
    fill_opacity=0.5,
    parse_html=False).add_to(map_toronto)
map_toronto

# If maps can not be seen on Github. Please check README file.
```

```python
# Using k Means clustering 
k=5
Tor_clustering = df_Toronto.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(Tor_clustering)
kmeans.labels_
df_Toronto.insert(0, 'Cluster Group', kmeans.labels_)
```

```python
# create map around Toronto
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df_Toronto['Latitude'], df_Toronto['Longitude'], df_Toronto['Neighbourhood'], df_Toronto['Cluster Group']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.5).add_to(map_clusters)
       
map_clusters
```
